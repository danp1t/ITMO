```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('sex_bmi_smokers.csv')
print(data.head())
```

```python
smoking_men = data[(data['sex'] == 'male') & (data['smoker'] == 'yes')].shape[0]
non_smoking_women = data[(data['sex'] == 'female') & (data['smoker'] == 'no')].shape[0]
```

```python
print("Количество курящих мужчин:", smoking_men)
print("Количество некурящих женщин:", non_smoking_women)
```

```python
print("Посчитаем выборочное среднее, выборочную дисперсию, медиану, квантиль 3/5 для всех людей")
summa = 0
count = 0
for i in data['bmi']:
    summa += i
    count += 1
average_bmi = summa / count
print("Выборочное среднее:", average_bmi)

summa = 0
for i in data['bmi']:
    summa += (i - average_bmi)**2
var_bmi = 1 / count * summa
print("Выборочная дисперсия:", var_bmi)

sorted_list = sorted(data['bmi'])
if len(sorted_list) % 2 == 1:
    print("Медиана:", sorted_list[len(sorted_list) // 2])
else:
    print("Медиана:", ((sorted_list[len(sorted_list) // 2]) + (sorted_list[len(sorted_list) // 2 - 1])) / 2)

print("Квантиль 3/5:", sorted_list[round(count*0.6) - 1])
```

```python
plt.figure(figsize=(18, 12))

plt.subplot(3, 2, 1)
sorted_bmi_all = np.sort(data['bmi'])
y_values_all = np.arange(1, len(sorted_bmi_all) + 1) / len(sorted_bmi_all)
plt.plot(sorted_bmi_all, y_values_all, label='Все наблюдатели')
plt.title('Эмпирическая функция распределения (Все)')
plt.xlabel('ИМТ')
plt.ylabel('Доля наблюдений')
plt.grid(True)
plt.legend()

plt.subplot(3, 2, 2)
plt.hist(data['bmi'], bins=30, color='blue', edgecolor='black', alpha=0.7, label='Все наблюдатели')
plt.title('Гистограмма ИМТ (Все)')
plt.xlabel('ИМТ')
plt.ylabel('Частота')
plt.grid(True)
plt.legend()

plt.subplot(3, 2, 3)
plt.boxplot(data['bmi'], vert=False, patch_artist=True, boxprops=dict(facecolor='orange'))
plt.title('Box-plot ИМТ (Все)')
plt.xlabel('ИМТ')
plt.grid(True)
```
